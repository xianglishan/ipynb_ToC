```python
import json
import re

def ipynb_toc(file_path):
    json_raw = open(file_path, 'r')
    json_dict = json.load(json_raw)
    json_raw.close()

    ## heading_list 見出しテキストのリスト
    ## toc_list 作る用
    ## json書き込む用に書き換えちゃう
    heading_list = []
    heading_num = 0
    for i in json_dict['cells']:
        if i['cell_type'] == 'markdown':
            heading_num_list =[]
            for n,t in enumerate(i['source']):
                if re.match("#+ .+\\n", t):
                    heading_list.append(t)
                    heading_num_list.append(n)
            for k,m in enumerate(heading_num_list):
                mm = m+k
                if m>=1 and re.match("<a id='.+'></a>\\n", i['source'][m-1]):
                    i['source'].pop(m-1)
                    mm-=1
                i['source'][mm:mm] = ['<a id="anchor_'+str(heading_num)+'"></a>\n']
                heading_num+=1

    ## toc_list ToC用のリスト
    ## heading_listから作る
    ## **ToC**って書いてあるマークダウンセルを書き換えちゃう
    toc_list = ['**ToC**\n']
    for n,i in enumerate(heading_list):
        t = re.sub(r'# ', "- [", i) + "]"
        t = re.sub(r'#', "    ", t)
        t = re.sub(r'\n', "", t)
        toc_list.append(t + '(#anchor_' + str(n) +')\n')

    for n,i in enumerate(json_dict['cells']):
        if i['cell_type']=='markdown' and i['source'][0]=='**ToC**':
            i['source'] = toc_list

    output = open(file_path,'w')
    json.dump(json_dict, output)
    output.close()
```

```python
# import re

# def toc_in_md(file_path)
file_path = './text.md'
out_path = './out.md'

f = open(file_path, 'r')
body = f.read()

heading_list = re.findall("#+ .+\\n", body)

for n,i in enumerate(heading_list):
    body = re.sub(i, '<a id="anchor'+str(n)+'"></a>\n'+i, body)

toc_list = []
for n, i in enumerate(heading_list):
    t = re.sub(r'# ', "- [", i) + "]"
    t = re.sub(r'#', "    ", t)
    t = re.sub(r'\n', "", t)
    toc_list.append(t + '(#anchor' + str(n) +')')

f.close()

f = open(out_path, 'w')
f.write(body)
f.close()

print('Insert the following text as toc')
print('\n'.join(toc_list))

# return '\n'.join(toc_list)
```
